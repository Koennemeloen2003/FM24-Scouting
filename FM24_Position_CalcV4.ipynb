The way it's set up the attributes are grouped from most to least important for a given role, this is done from top to bottom, in this example the top 4 atributes are the most important and they have a weight of 5. The middle block has a weight of 3 and the bottom one a weight of 1.

So to change the importance given to a certain attribute all you need to do is change what block it is in, if you want to make it more important you move it up a block and vice versa for making a attribute less important.

Keep in mind that in the end everyting gets devided by the total weight, in this example 51, this is just counting howmany attributes there are per block and than multiplying that with the weight of th block.

E.G.:
1. We have 4 attributes in the top box wich is multiplied by 5
2. we have 7 attributes in the middle box which is multiplied by 3
3. We have 10 attributes in the bottom box which is multiplied by 1

so our devider will be 4X5+7X3+10X1=51
